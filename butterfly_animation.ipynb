```python
from IPython.display import HTML, display

html_code = """
<canvas id="butterfly-canvas" width="700" height="480" style="background: #0d1117; border-radius: 12px;"></canvas>
<script>
(function() {
    const canvas = document.getElementById('butterfly-canvas');
    const ctx = canvas.getContext('2d');
    const W = canvas.width, H = canvas.height;

    const palettes = [
        ['#F0997B', '#D85A30'],
        ['#7F77DD', '#534AB7'],
        ['#5DCAA5', '#0F6E56'],
        ['#ED93B1', '#993556'],
        ['#FAC775', '#BA7517']
    ];

    class Butterfly {
        constructor() {
            this.baseX = Math.random() * W;
            this.baseY = Math.random() * H;
            this.angle = Math.random() * Math.PI * 2;
            this.radius = 40 + Math.random() * 60;
            this.driftSpeed = 0.02 + Math.random() * 0.02;
            this.flapPhase = Math.random() * Math.PI * 2;
            this.flapSpeed = 0.15 + Math.random() * 0.1;
            this.size = 10 + Math.random() * 8;
            const pal = palettes[Math.floor(Math.random() * palettes.length)];
            this.colorLight = pal[0];
            this.colorDark = pal[1];
            this.driftAngleOffset = Math.random() * Math.PI * 2;
            this.x = this.baseX;
            this.y = this.baseY;
        }

        update(t) {
            this.angle += this.driftSpeed;
            this.baseX += Math.cos(t * 0.4 + this.driftAngleOffset) * 0.3;
            this.baseY += Math.sin(t * 0.3 + this.driftAngleOffset) * 0.2;

            if (this.baseX < -50) this.baseX = W + 50;
            if (this.baseX > W + 50) this.baseX = -50;
            if (this.baseY < -50) this.baseY = H + 50;
            if (this.baseY > H + 50) this.baseY = -50;

            this.x = this.baseX + Math.cos(this.angle) * this.radius;
            this.y = this.baseY + Math.sin(this.angle) * this.radius * 0.6;

            this.flapPhase += this.flapSpeed;
        }

        draw(ctx) {
            const flap = Math.sin(this.flapPhase);
            const wingSpread = 0.3 + Math.abs(flap) * 0.7;
            const moveDir = this.angle + Math.PI / 2;

            ctx.save();
            ctx.translate(this.x, this.y);
            ctx.rotate(moveDir);

            const grad = ctx.createLinearGradient(0, -this.size, 0, this.size);
            grad.addColorStop(0, this.colorLight);
            grad.addColorStop(1, this.colorDark);
            ctx.fillStyle = grad;
            ctx.globalAlpha = 0.9;

            [-1, 1].forEach(side => {
                ctx.save();
                ctx.scale(wingSpread, 1);
                ctx.beginPath();
                ctx.ellipse(side * this.size * 0.6, -this.size * 0.3, this.size * 0.7, this.size * 0.5, 0, 0, Math.PI * 2);
                ctx.fill();
                ctx.beginPath();
                ctx.ellipse(side * this.size * 0.5, this.size * 0.35, this.size * 0.5, this.size * 0.35, 0, 0, Math.PI * 2);
                ctx.fill();
                ctx.restore();
            });

            ctx.globalAlpha = 1;
            ctx.fillStyle = this.colorDark;
            ctx.fillRect(-1, -this.size * 0.5, 2, this.size);

            ctx.restore();
        }
    }

    const butterflies = [];
    for (let i = 0; i < 14; i++) butterflies.push(new Butterfly());

    let t = 0;
    function animate() {
        ctx.fillStyle = 'rgba(13, 17, 23, 0.25)';
        ctx.fillRect(0, 0, W, H);

        t += 0.016;
        butterflies.forEach(b => {
            b.update(t);
            b.draw(ctx);
        });

        requestAnimationFrame(animate);
    }
    animate();
})();
</script>
"""

display(HTML(html_code))
```
