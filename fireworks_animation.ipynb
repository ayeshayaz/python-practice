```python
from IPython.display import HTML, display

html_code = """
<canvas id="fireworks-canvas" width="700" height="450" style="background: linear-gradient(135deg, #0a0a14 0%, #1a1a2e 100%); border: 2px solid #4ECDC4; border-radius: 8px;"></canvas>

<script>
(function() {
    const canvas = document.getElementById('fireworks-canvas');
    const ctx = canvas.getContext('2d');
    const colors = ['#FF6B6B', '#4ECDC4', '#FFE66D', '#95E1D3', '#F38181', '#C7CEEA'];

    class Particle {
        constructor(x, y) {
            this.x = x;
            this.y = y;
            this.vx = (Math.random() - 0.5) * 12;
            this.vy = (Math.random() - 0.5) * 12;
            this.life = 1.0;
            this.size = Math.random() * 5 + 3;
            this.color = colors[Math.floor(Math.random() * colors.length)];
        }
        update() {
            this.vy += 0.2;   // gravity
            this.x += this.vx;
            this.y += this.vy;
            this.vx *= 0.98;  // air resistance
            this.life -= 0.015;
        }
        draw(ctx) {
            ctx.save();
            ctx.globalAlpha = Math.max(this.life, 0);
            ctx.fillStyle = this.color;
            ctx.beginPath();
            ctx.arc(this.x, this.y, this.size, 0, Math.PI * 2);
            ctx.fill();
            ctx.restore();
        }
    }

    let particles = [];
    let frameCount = 0;

    function createBurst(x, y, count = 80) {
        for (let i = 0; i < count; i++) particles.push(new Particle(x, y));
    }

    function animate() {
        if (frameCount % 60 === 0) {
            const x = Math.random() * (canvas.width - 200) + 100;
            const y = Math.random() * 250 + 50;
            createBurst(x, y);
        }

        ctx.fillStyle = 'rgba(10, 10, 20, 0.15)';
        ctx.fillRect(0, 0, canvas.width, canvas.height);

        // Update, draw, AND remove dead particles — this is the key fix
        for (let i = particles.length - 1; i >= 0; i--) {
            particles[i].update();
            if (particles[i].life <= 0) {
                particles.splice(i, 1);
            } else {
                particles[i].draw(ctx);
            }
        }

        frameCount++;
        requestAnimationFrame(animate);
    }

    createBurst(canvas.width / 2, 200);
    animate();
})();
</script>
"""

display(HTML(html_code))
```
